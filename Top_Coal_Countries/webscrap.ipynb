```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_coal_production'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table = soup.find('table', class_="wikitable sortable")
```

```python
titles = table.find_all("th")
```

```python
table_title = [title.text.strip() for title in titles]

print(table_title)
```

```python
cleaned_title = [item.split('[')[0] for item in table_title]

print(cleaned_title)
```

```python
cleaned_title.pop(11)
```

```python
cleaned_title
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=cleaned_title)
```

```python
df
```

```python
#Let's find out the data from the table
table_list = table.find_all('tr')
table_list
```

```python

for row in table_list[1:]:
    data = row.find_all('td')
    row_data = [data.text.strip() for data in data]

    df.loc[len(df)] = row_data

    
```
